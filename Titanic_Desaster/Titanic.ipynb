### Step 1. Import the necessary libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv',sep=',')
```

### Step 3. Assign it to a variable titanic 

```python
titanic = df.copy()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId',inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex1 = lambda x: 1 if x =='male' else 0
titanic['sex1'] = titanic.Sex.apply(sex1)
```

```python
male = sum(titanic.sex1 == 1)
female = sum(titanic.sex1 == 0)
female
```

```python
y = np.array([314,577])
gender =['female','male']
plt.pie(y, labels = gender)
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
%matplotlib notebook

plt.scatter(  titanic.Age ,titanic.Fare, c=titanic.sex1 )
plt.legend(['Male'])
plt.show()
```
